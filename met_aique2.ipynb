```python
import pandas as pd
import re
import numpy as np
```

```python
maq=pd.read_excel('met_aique.xlsx')
```

```python
maq[maq['ilustrador'].notnull()]['ilustrador']
```

```python
for i in maq[maq['ilustrador'].notnull()].index:
    isb=maq['ilustrador'].iloc[i]
    ilu=re.search(r'>(\w* ?\w*?)</a>',isb)
    if ilu != None:
        maq['ilustrador'].iat[i]=ilu.group(1)
```

```python
maq['paginas']=maq['paginas'].apply(lambda x: str(x).replace(" páginas",""))
```

```python
maq['titulo']=maq['titulo'].apply(lambda x: str(x).replace("\n",""))
```

```python
maq.coleccion.unique()
```

```python
tabla_formato={
    '  Ser Director ':("Trade","Rústica"),
    '“El abecé de…”':("Trade","Rústica"),
    'Aportes a la   Educación Inicial':("Trade","Rústica"),
    'Biblioteca  esencial':("Trade","Rústica"),
    'Carrera docente':("Trade","Rústica"),
    'Escuela de Educación  Universidad de San Andrés (UDESA) ':("Trade","Rústica"),
    'Herramientas  para...':("Bolsillo","Rústica"),
    'La trama que trama':("Bolsillo","Rústica"),
    'Neurociencias   y Educación ':("Trade","Rústica"),
    'Nueva carrera   docente ':("Trade","Rústica"),
    'Política y  Educación ':("Trade","Rústica"),
    'Psicología Cognitiva y Educación':("Trade","Rústica"),
    'Serie de la  palabra':("Trade","Rústica"),
    'Sobrelecturas':("Trade","Rústica"),
    'Sobrelecturas  Directora: Elena Luchetti ':("Trade","Rústica"),
    np.NaN:("Trade","Rústica")
}
```

```python
maq['formato_1']=maq['coleccion'].apply(lambda x: tabla_formato[x][0])
maq['formato_2']=maq['coleccion'].apply(lambda x: tabla_formato[x][1])
maq['bolsillo']=maq['coleccion'].apply(lambda x: tabla_formato[x][0])
```
